## Data Preprocessing and Feature Engineering

Handling missing values

```python
data.isnull().sum()
```

As, last five columns have more than 95% of them are null values,we will ignore them.

For particular AGENCY code there is unique AGENCY_NAME so, we are dropping AGENCY.

We are droping LAST NAME,FIRST NAME,MI since there is no relation with target variable.

For particular CLASS_TITLE there is unique CLASS_CODE so, we are dropping CLASS_CODE.

```python
data.columns
```

```python
data=data.drop(['AGENCY','LAST NAME','FIRST NAME','MI','CLASS_CODE','duplicated','multiple_full_time_jobs','combined_multiple_jobs','summed_annual_salary','hide_from_search'],axis=1)
```

```python
data
```

## Categorical Encoding

```python
#  Converting object dtype date to datetime dtype
from datetime import datetime as dt
data['EMPLOY_DATE'] = pd.to_datetime(data['EMPLOY_DATE'])
# add a column for Year
data['Year'] = data['EMPLOY_DATE'].dt.year
```

```python
data
```

```python
# Fetching today's date and removing time stamp(converting time to zeros)
import datetime as dt
data['cur_yr']=dt.datetime.combine(dt.date.today(), dt.datetime.min.time())
#data['cur_yr']= dt.datetime(date.today())
```

```python
data
```

```python
#  We are subtracting Current year with EMPLOY_DATE and applying it to lambda function to calculate years for each record.
data['yrs']=((data.cur_yr)-(data.EMPLOY_DATE)).apply(lambda x: str(x).split(' ')[0]).apply(lambda x: int(x)/365)
```

```python
data
```

```python
data.yrs.max()
```

```python
data.yrs.min()
```

```python
neg= data.loc[data['yrs']<0]
neg
```

```python
# The joining date of employee is higher than the Current Date which leads to negative years of experiance so, we are dropping those 5 records.
data.drop(neg.index,inplace=True)
```

```python
data
```

```python
data.drop(['EMPLOY_DATE','Year','cur_yr'],inplace=True,axis=1)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
lc=LabelEncoder()
data.AGENCY_NAME=lc.fit_transform(data.AGENCY_NAME)
data.CLASS_TITLE=lc.fit_transform(data.CLASS_TITLE)	
data.ETHNICITY=lc.fit_transform(data.ETHNICITY)
data.GENDER=lc.fit_transform(data.GENDER)
data.STATUS=lc.fit_transform(data.STATUS)
```
